## Imports & Load the Model

```python
import joblib
import pandas as pd
import json

# Load the trained model
model = joblib.load("model_rf.joblib")
```

# Define a Simple Inference Function

```python
def predict_late_payment(input_dict):
    """
    input_dict: dict with keys matching feature names
    returns: prediction label (0 or 1)
    """
    # Ensure the input data has the same order as model expects
    input_dict_ordered = {key: input_dict[key] for key in model.feature_names_in_}
    
    # Create a DataFrame with the ordered input
    input_df = pd.DataFrame([input_dict_ordered])
    
    # Predict using the model
    prediction = model.predict(input_df)
    
    # Return the prediction as an integer (0 or 1)
    return int(prediction[0])
```
